```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import glob
import json
from pathlib import Path
```

```python
files = glob.glob('Resources/*.csv')
frame = [] 
for file in files:
    frame.append(pd.read_csv(file))
    #combined_df = pd.read_csv(file)
    
combined_df = pd.concat(frame)
year2016 = combined_df[combined_df['DATA_YEAR'] == 2016]
year2017 = combined_df[combined_df['DATA_YEAR'] == 2017]
year2018 = combined_df[combined_df['DATA_YEAR'] == 2018]
year2019 = combined_df[combined_df['DATA_YEAR'] == 2019]
year2020 = combined_df[combined_df['DATA_YEAR'] == 2020]
```

```python
jancount2016 = year2016['MONTH_NUM'].value_counts()[1]
febcount2016 = year2016['MONTH_NUM'].value_counts()[2]
marcount2016 = year2016['MONTH_NUM'].value_counts()[3]
aprcount2016 = year2016['MONTH_NUM'].value_counts()[4]
maycount2016 = year2016['MONTH_NUM'].value_counts()[5]
juncount2016 = year2016['MONTH_NUM'].value_counts()[6]
julcount2016 = year2016['MONTH_NUM'].value_counts()[7]
augcount2016 = year2016['MONTH_NUM'].value_counts()[8]
sepcount2016 = year2016['MONTH_NUM'].value_counts()[9]
octcount2016 = year2016['MONTH_NUM'].value_counts()[10]
novcount2016 = year2016['MONTH_NUM'].value_counts()[11]
deccount2016 = year2016['MONTH_NUM'].value_counts()[12]
```

```python
months = ["January", "Febuary", "March", "April", "May", "June", "July", "August", "September", "October", "Novemember", "December"]
crime2016 = [jancount2016, febcount2016, marcount2016, aprcount2016, maycount2016, juncount2016, julcount2016, augcount2016, sepcount2016, octcount2016, novcount2016, deccount2016]  
```

```python
x_axis = np.arange(len(crime2016))
plt.xlabel("Month")
plt.ylabel("Incidents Reported")
plt.title("Incidents Reported 2016")


tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, months, rotation = 45)
plt.bar(x_axis, crime2016, color="b", align="center")
```

```python
janavg = jancount / 5
febavg = febcount / 5
maravg = marcount / 5
apravg = aprcount / 5
mayavg = maycount / 5
junavg = juncount / 5
julavg = julcount / 5
augavg = augcount / 5
sepavg = sepcount / 5
octavg = octcount / 5
novavg = novcount / 5
decavg = deccount / 5
```

```python
avg_crime = [janavg, febavg, maravg, apravg, mayavg, junavg, julavg, augavg, sepavg, octavg, novavg, decavg]
x_axis = np.arange(len(avg_crime))
```

```python
plt.xlabel("Month")
plt.ylabel("Incidents Reported")
plt.title("Average Incidents Reported Per Month 2016-2020")


tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, months, rotation = 45)
plt.bar(x_axis, avg_crime, color="r", align="center")
```

```python
janfeb = (febavg-janavg)/janavg *100
febmar = (maravg-febavg)/febavg *100
marapr = (apravg-maravg)/maravg *100
aprmay = (mayavg-apravg)/apravg *100
mayjun = (junavg-mayavg)/mayavg *100
junjul = (julavg-junavg)/junavg *100
julaug = (augavg-julavg)/julavg *100
augsep = (sepavg-augavg)/augavg *100
sepoct = (octavg-sepavg)/sepavg *100
octnov = (novavg-octavg)/octavg *100
novdec = (decavg-novavg)/novavg *100
decjan = (janavg-decavg)/decavg *100
```
