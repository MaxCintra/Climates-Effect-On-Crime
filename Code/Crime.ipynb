```python
janavg = jancount / 5
febavg = febcount / 5
maravg = marcount / 5
apravg = aprcount / 5
mayavg = maycount / 5
junavg = juncount / 5
julavg = julcount / 5
augavg = augcount / 5
sepavg = sepcount / 5
octavg = octcount / 5
novavg = novcount / 5
decavg = deccount / 5
```

```python
avg_crime = [janavg, febavg, maravg, apravg, mayavg, junavg, julavg, augavg, sepavg, octavg, novavg, decavg]
x_axis = np.arange(len(avg_crime))
```

```python
plt.xlabel("Month")
plt.ylabel("Incidents Reported")
plt.title("Average Incidents Reported Per Month 2016-2020")


tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, months, rotation = 45)
plt.bar(x_axis, avg_crime, color="r", align="center")
```

```python
janfeb = round((febavg-janavg)/janavg *100, 2)
febmar = round((maravg-febavg)/febavg *100, 2) 
marapr = round((apravg-maravg)/maravg *100, 2) 
aprmay = round((mayavg-apravg)/apravg *100, 2) 
mayjun = round((junavg-mayavg)/mayavg *100, 2) 
junjul = round((julavg-junavg)/junavg *100, 2) 
julaug = round((augavg-julavg)/julavg *100, 2) 
augsep = round((sepavg-augavg)/augavg *100, 2) 
sepoct = round((octavg-sepavg)/sepavg *100, 2) 
octnov = round((novavg-octavg)/octavg *100, 2) 
novdec = round((decavg-novavg)/novavg *100, 2) 
decjan = round((janavg-decavg)/decavg *100, 2) 
```

```python
month_month =["December to january", "January to Febuary", "Febuary to March", "March to April", "April to May", "May to June", "June to July", "July to August", "August to Septmeber", "September to October", "October to November", "November to December"] 
change = [decjan, janfeb, febmar, marapr, aprmay, mayjun, junjul, julaug, augsep, sepoct, octnov, novdec]
```

```python
plt.xlabel("Month")
plt.ylabel("% Change")
x_axis = np.arange(len(change))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, month_month, rotation = 90)
p1 = plt.bar(x_axis, change)
for rect1 in p1:
    height = rect1.get_height()
    plt.annotate( "{}%".format(height),(rect1.get_x() + rect1.get_width()/2, height),ha="center",va="bottom",fontsize=10, ) 
plt.plot
```
