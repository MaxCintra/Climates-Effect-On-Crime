```python
import hvplot.pandas
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from scipy.stats import linregress

weather_data_df = pd.read_csv("weather_data/average_temp_by_state.csv")

weather_data_df
```

```python
# Cleaning up the dataframe and sortin the data

filtered_weather_df = weather_data_df

filtered_weather_df = filtered_weather_df.loc[(filtered_weather_df["year"] >= 2016) & (filtered_weather_df["year"] <= 2021)]

filtered_weather_df = filtered_weather_df.loc[(filtered_weather_df["state"] == "Kansas")]

filtered_weather_df = filtered_weather_df.drop(columns=["Unnamed: 0","monthly_mean_from_1901_to_2000","centroid_lon","centroid_lat"])

filtered_weather_df
```

```python
#Code thats used for pretty much most of the bar charts.

monthly_mean_temps = filtered_weather_df.groupby("month")["average_temp"].mean()

year_to_plot = 2016
monthly_temps = filtered_weather_df.loc[filtered_weather_df["year"] == year_to_plot]

monthly_mean_temps = monthly_temps.groupby("month")["average_temp"].mean()

x_axis = monthly_mean_temps.index.values
y_axis = monthly_mean_temps.values

plt.bar(x_axis, y_axis, align="center", color="cyan")

plt.xlabel("Month")
plt.ylabel("Temperature (Â°F)")
plt.title(f"Monthly Mean Temperatures for {year_to_plot}")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"], rotation="vertical")
plt.show()

plt.show()
```
